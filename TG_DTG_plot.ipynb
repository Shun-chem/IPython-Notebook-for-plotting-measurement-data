```python
# Inport of necessary modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# Input of data
input('Copy temperature data to Clipboad and then input Enter')
Temp = pd.read_clipboard()
input('Copy TG/mg data to Clipboad and then input Enter')
TG = pd.read_clipboard()
input('Copy DTG/mgs-1 data to Clipboad and then input Enter')
DTG = pd.read_clipboard()

w0 = float(input('Input initial weight(mg)'))
```

```python
# Data process
TG_net = TG + w0
TG_net_ratio = TG_net / w0 * 100
```

```python
fig, ax1 = plt.subplots(figsize=(15,10))
ax1.plot(Temp,TG_net_ratio, 
         marker='.',linewidth=0.5, 
         label='TG',
         markersize=0.5, 
         color='red')
ax2 = ax1.twinx()  # 2つのプロットを関連付ける
ax2.plot(Temp,DTG, 
         marker='.',linewidth=1, 
         label='DTG',
         markersize=0.5, 
         color='blue')
x0 = np.arange(20,501,10)
#ax1
ax1.set_xticks(x0)
ax1.set_xlim(20,500)
ax1.set_xticklabels(x0,rotation=90)
ax1.set_xlabel('Temperature (degC)',fontsize = 20)
ax1.set_ylim(65,101)
ax1.grid()
ax1.set_yticks(np.arange(60,101,1))
ax1.set_ylabel('w/w0 (%)', color='r',fontsize = 20)
for tl in ax1.get_yticklabels():
    tl.set_color('r')
#ax2
ax2.set_xticks(x0)
ax2.set_xticklabels(x0,rotation=90)
ax2.set_ylabel('DTG (mg/s)', color='b',fontsize = 20)
for tl in ax2.get_yticklabels():
    tl.set_color('b')

plt.subplots_adjust(hspace=0.7,bottom=0.2)
ax1.tick_params(labelsize = 10)
plt.title('TG / DTG', fontsize=30) #Title
plt.show()
```
